```python
class Config:
    epochs = 15
    imgsz = 384
    deivce = 'cuda' if torch.cuda.is_available() else 'CPU'
    batch_size = 32
```

```python
model = YOLO("yolo11l.pt")

results = model.train(data="asl.yaml",
                      epochs=Config.epochs,
                      imgsz=Config.imgsz,
                      device=Config.deivce)
```

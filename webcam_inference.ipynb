```python
import cv2
from ultralytics import YOLO
```

```python
metrics_files = {
    'YOLO-n': './runs/detect/train6',
    'YOLO-l': './runs/detect/train15',
    'YOLO-x': './runs/detect/train_yolo_x',
    'RT-DETR-l': '/runs/detect/train14',
    'RT-DETR-x': '/runs/detect/train_rtdetr_x',
}
```

```python
yolo11l_model = YOLO(f'./{metrics_files["YOLO-x"]}/weights/best.pt')
```

```python
# Open the webcam (use 0 for default webcam, or replace with your camera index)
cap = cv2.VideoCapture(0)

# Check if the webcam is opened successfully
if not cap.isOpened():
    print("Error: Cannot access webcam.")
    exit()

# Process frames from the webcam
while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break

    # Run YOLO model on the frame
    results = yolo11l_model(frame)

    # Visualize the predictions on the frame
    annotated_frame = results[0].plot()

    # Display the annotated frame
    cv2.imshow("YOLOv11 Object Detection", annotated_frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam and close OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
